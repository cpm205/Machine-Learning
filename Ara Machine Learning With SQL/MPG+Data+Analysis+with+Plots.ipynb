### Python Code and graphic examples 

###### load the library Matplotlib, a common libary for creating graphics in Python

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#load the Dataset
data = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data-original",
                   delim_whitespace = True, header=None,
                   names = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration',
                            'model', 'origin', 'car_name'])

       
data.plot.scatter('displacement', 'mpg')   
```

```python
import scipy.optimize as sc
import numpy as np
```

```python
print(data.shape)
data = data.dropna()
data.head()
```

```python
import seaborn as sns

```

```python
p = plt.hist(data.mpg)
plt.title("MPG")
p
sns.despine()
```

```python
sns.lmplot("mpg", "weight", data);
```

```python
sns.lmplot("mpg", "weight", data, order=2);
```

```python
sns.jointplot("mpg", "weight", data, kind="reg")
```
